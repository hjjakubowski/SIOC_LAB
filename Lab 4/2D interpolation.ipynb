```python
import numpy as np
from numpy.typing import NDArray
from skimage import io, transform, color
from src.kernels import sample_hold_kernel, nearest_neighbour_kernel, linear_kernel
from src.interp import image_interpolate2d, image_interpolate2d_rgb
import matplotlib.pyplot as plt
```

```python
grey_img = io.imread('image.png')
grey_img = transform.resize(grey_img, output_shape=(100,100))
io.imshow(grey_img)
grey_img.shape
```

```python
hn_interp = image_interpolate2d(grey_img, ratio=4, kernel_function=nearest_neighbour_kernel)
hs_interp = image_interpolate2d(grey_img, ratio=4, kernel_function=sample_hold_kernel)
hl_interp = image_interpolate2d(grey_img, ratio=4, kernel_function= linear_kernel)
plt.figure(figsize=(18,6))

plt.subplot(1, 3, 1)  
plt.imshow(hn_interp, cmap="gray") 
plt.title("Nearest Neighbour Kernel")


plt.subplot(1, 3, 2) 
plt.imshow(hs_interp, cmap="gray") 
plt.title("Sample Hold Kernel")

plt.subplot(1, 3, 3) 
plt.imshow(hl_interp, cmap="gray") 
plt.title("Linear Kernel")
#Zmiana wartoci zwracanej w funkcji interpolujcej pozwolia zej z czasem oblicze z ok 6 min do ok 1min 30s
```
